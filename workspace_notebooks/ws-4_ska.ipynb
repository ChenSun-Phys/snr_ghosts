### Deriving the efficiency

the Bose distribution correction to temperature

```python
def corr_fn(nu, T):
    x = 2.*np.pi * nu / T * ct._GHz_over_eV_ / ct._K_over_eV_
    res = x / (np.exp(x) - 1.)
    return res
    
corr_fn(10, 20)

# the atm opacity is also a small correction
```
