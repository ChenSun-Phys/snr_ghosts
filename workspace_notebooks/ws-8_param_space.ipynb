### 2.2 $(L_{\rm peak}, t_{\rm peak})$ parameter space: **Analytic**

```python
def nu_dependence_fn(nu, sn_remnant):
    
    # compute a mock approximation to the solid angle of the signal
    Omega_size = sn_remnant.get_size()
    Omega_dispersion = ct.angle_to_solid_angle(2.*(1.e-3 / 2.17)) # true for x_wavefront < distance: this is the smallest possible dispersion
    mock_Omega_signal = max(Omega_size, Omega_dispersion) # the smallest possible signal: Omega_dispersion and Omega_aberration (and thus Omega_signal) could be larger
    
    ma = pt.ma_from_nu(nu) # axion mass [eV]
    area, window, Tr, Omega_res = rt.SKA_rescaled_specs(ma, data={'exper':'SKA'}) # SKA specs
    
    mock_Omega_obs = np.maximum.reduce([mock_Omega_signal*np.ones_like(Omega_res), Omega_res]) # mock observation solid angle
    
    print (Omega_res > mock_Omega_signal)
    print Tr
    
    l, b = sn_remnant.get_coord() # SNR coordinates
    l_echo, b_echo = l + 180., -b # echo coordinates
    
    Tbg_408 = ap.bg_408_temp(l_echo, b_echo, size=mock_Omega_obs, average=True) # bg temperature at 408 MHz
    Tn = ap.T_noise(nu, Tbg_at_408=Tbg_408, beta=-2.55, Tr=Tr) # noise temperature at freq. nu
    
    alpha = sn_remnant.alpha
    f = nu**(-alpha-0.5) * area / Tn / sqrt(mock_Omega_obs/Omega_res)
    
    return f
```
